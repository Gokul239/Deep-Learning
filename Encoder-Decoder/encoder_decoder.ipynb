```python
# Function to build the vocabulary

def build_vocab(data):
    tokens =  Counter()
    for sent in data:
        tokens.update(sent.split())
    vocab = {"<PAD>": 0, "<SOS>": 1, "<EOS>": 2}
    for i, token in enumerate(tokens.keys(), 3):
        vocab[token] = i
    return vocab
```

```python
# Sentence to indices 

def sent2indices(data, vocab):
    indices = [vocab.get(i.lower(), vocab.get('<UNK>', 0)) for i in data.split()]
    data_indices = [vocab['<SOS>']] + indices + [vocab['<EOS>']]
    return data_indices
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Preprocessing the text, converting the text into sequences with padding
def pre_processing(sequence_data, src_vocab, targ_vocab):
    src_data = [sent2indices(data[1].lower(), src_vocab) for data in sequence_data]
    tar_data = [sent2indices(data[0], targ_vocab) for data in sequence_data]

    # padding to the dataset

    src_seq = pad_sequences(src_data, padding='post', value= src_vocab['<PAD>'])
    tar_seq = pad_sequences(tar_data, padding='post', value= targ_vocab['<PAD>'])

    return(src_seq, tar_seq)
```

```python
eng_seq, tam_seq = pre_processing(sentence_pairs, eng_vocab, tam_vocab)
```

```python
# Creating an encoder
from tensorflow.keras.layers import Embedding, LSTM, Dense

class Encoder(tf.keras.Model):
    def __init__(self, input_size, embed_size, hidden_size):
        super(Encoder, self).__init__()

        self.embedding = Embedding(input_size, embed_size)
        self.lstm = LSTM(hidden_size, return_state=True)

    def call(self, data):
        embed = self.embedding(data)
        _, hidden, cell = self.lstm(embed)
        return hidden, cell

# Creating decoder    
class Decoder(tf.keras.Model):
    def __init__(self, output_size, embed_size, hidden_size):
        super(Decoder, self).__init__()
        self.embedding = Embedding(output_size, embed_size)
        self.lstm = LSTM(hidden_size, return_sequences=True, return_state=True)
        self.dense = Dense(output_size)

    def call(self, data, hidden, cell):
        embed = self.embedding(data)
        lstm_out, hc, fc = self.lstm(embed, initial_state=[hidden, cell])
        output = self.dense(lstm_out)

        return output, hc, fc
```

```python
# Creating an sequence to sequennce class using encoder and decoder

class Seq2seq(tf.keras.Model):
    def __init__(self, encoder, decoder):
        super(Seq2seq, self).__init__()
        self.encoder = encoder
        self.decoder = decoder

    def call(self, inputs, teacher_forcing_ratio = 1.0):
        src_data, targ_data = inputs
        targ_len = targ_data.shape[1]

        # Encoder

        hidden, cell = self.encoder(src_data)

        output_lst = []

        inputs = targ_data[:, 0:1]

        for i in range(1, targ_len):
            output, hidden, cell = self.decoder(inputs, hidden, cell)
            output_lst.append(output)

            inputs= targ_data[:,i:i+1] if random.random()>= teacher_forcing_ratio else tf.argmax(output, axis=-1, output_type=tf.int32)
        return tf.concat(output_lst, axis=1)
```

```python
# Create a training function

def train(model, src_data, targ_data, epochs=50):
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)    #optimizer
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction='none')    #loss

    def train_step(src, tgt):

        with tf.GradientTape() as gtape:
            output = model([src, tgt])
            targ_labels = tgt[:, 1:]
            # print(tf.shape(targ_labels)[1])
            output = output[:, :tf.shape(tgt)[1], : ]

            mask = tf.cast(targ_labels != tam_vocab['<PAD>'], tf.float32)

            # loss
            loss = loss_fn(targ_labels, output)
            loss*=mask
            total_loss = tf.reduce_sum(loss)
            tokens = tf.reduce_sum(mask)
            mean_loss = total_loss/(tokens+1e-8)

        gradients = gtape.gradient(mean_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))

        return mean_loss            

    for epoch in range(epochs):

        loss = train_step(src_data, targ_data)

        if epoch%10 == 0:
            print(f'Epochs: {epoch}, Loss: {loss.numpy():.2f}')

    print("Training Completed")
```

```python
input_size = len(eng_vocab)
hidden_size = 32
embed_size = 10
output_size= len(tam_vocab)

encoder = Encoder(input_size, embed_size, hidden_size)
decoder = Decoder(output_size, embed_size, hidden_size)
s2s = Seq2seq(encoder, decoder)
```
